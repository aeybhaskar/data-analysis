## Cleaning the data set by dropping irrelevant Columns

```python
dataframe = df.drop(['approxdate' , 'extended','resolution', 'latitude' ,'attacktype2','attacktype2_txt','attacktype3','attacktype3_txt','longitude' ,'specificity','guncertain2','guncertain3','vicinity','location','summary','crit1','crit2','crit3','doubtterr' ,'alternative', 'alternative_txt' , 'ransom','ransomamt','ransomamtus','ransompaid','ransompaidus','ransomnote','hostkidoutcome','hostkidoutcome_txt','nreleased','addnotes','scite1','scite2' , 'scite3' ,'nwound','nwoundus','nwoundte', 'propcomment','ishostkid' ,'nhostkid','nhostkidus','nhours','ndays','divert','kidhijcountry','dbsource' , 'nperpcap', 'claimed','claimmode','claimmode_txt','claim2','claimmode2','claimmode2_txt','claim3','claimmode3','claimmode3_txt','compclaim'], axis =1)
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(dataframe.isnull(),cbar = False, cmap = 'viridis')
```

## Analysing and finding relationships in the given dataset

```python
plt.figure(figsize=(60,60))
sns.barplot(x='region_txt', y='region', data = dataframe)
```

The above plot is a bar graph plotted between Region and Region_txt and it shows all the regions coressponding to the number assigned to them in the database.

```python
plt.figure(figsize=(20,20))
sns.barplot(x='attacktype1_txt', y = 'attacktype1', data = dataframe)
```

The above plot is a bar graph plotted between Attacktype1 and Attacktype1_txt and it shows all the Attacktypes corresponding to the number assigned to them in the database.

```python
plt.figure(figsize=(60,60))
plt.axis([0,15,1950,2000])
sns.barplot(x='region' , y='iyear' ,hue='attacktype1_txt' , data=dataframe)
```


The above bar graph plotted between Region and Year with hue as attacktype1_txt. This will help us know the frequency of attacks in a specific region and how much are they prevelant. Helping us know the hot zones of terrorism.



```python
plt.figure(figsize=(30,30))
sns.countplot(x='multiple', hue = 'region_txt' , data=dataframe)
```

```python
plt.figure(figsize=(20,20))
sns.countplot(x='multiple', hue = 'attacktype1_txt' , data=dataframe)
```

From the above graphs , following can be inferred :

Multiple attacks have not been reported in many regions. Regions where they have been reported are North America , Western Europe and Middle East and north Africa. North America having a higher number.

Out of these multiple attacks , bombing and explosion are more frequent , then come facility/infrastructure attack , then is hijacking , Armed assaults , followed by hostage taking.

From the plot between iyear and regions , it can be seen that the hot zone of terrorism are North America , Middle East and North Africa , south America and Western Europe.
